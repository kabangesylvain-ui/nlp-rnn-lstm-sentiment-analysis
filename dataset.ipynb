# Téléchargement et exploration du dataset IMDb

## Objectif

L'objectif de ce notebook est de télécharger le dataset **IMDb Movie Reviews** depuis HuggingFace, de le sauvegarder localement en format CSV, et de vérifier sa structure pour les étapes ultérieures du projet (prétraitement, entraînement, évaluation).

---

## Pourquoi ce dataset ?

Le dataset IMDb est un **standard en NLP** pour la classification de sentiments car :

| Critère | Description |
|---------|-------------|
| **Taille** | 50 000 critiques (25K train / 25K test) |
| **Équilibre** | 50% positifs / 50% négatifs |
| **Qualité** | Annotations manuelles fiables |
| **Accessibilité** | Disponible publiquement sur HuggingFace |
| **Longueur** | Textes de taille réelle (moyenne 234 mots) |

Ce dataset permet une **comparaison fiable** entre RNN et LSTM.

### Étape 2 Téléchargement du dataset IMDb

```python
# Téléchargement du dataset IMDb
from datasets import load_dataset

print("📥 Téléchargement en cours ")
 
dataset = load_dataset("imdb")

print(f"✅ Téléchargé !")
print(f"   Train : {len(dataset['train']):,} exemples")
print(f"   Test  : {len(dataset['test']):,} exemples")
```

### Étape 3. Sauvegarde en CSV 

```python
# Sauvegarde en CSV locaux
import pandas as pd
import os

os.makedirs("data", exist_ok=True)

train_df = pd.DataFrame(dataset["train"])
test_df  = pd.DataFrame(dataset["test"])

train_df.to_csv("data/imdb_train.csv", index=False)
test_df.to_csv("data/imdb_test.csv",   index=False)

print("✅ Fichiers sauvegardés :")
print(f"   data/imdb_train.csv  → {len(train_df):,} lignes")
print(f"   data/imdb_test.csv   → {len(test_df):,} lignes")
```

### Étape 4 Vérification

```python
# Vérification
train_check = pd.read_csv("data/imdb_train.csv")

print("=== VÉRIFICATION ===")
print(f"Shape : {train_check.shape}  ← doit être (25000, 2)")
print()
print(train_check.head(3))
print()
print("Labels :")
print(train_check["label"].value_counts())
print()
print("✅ TOUT OK — Dataset prêt !")
```
